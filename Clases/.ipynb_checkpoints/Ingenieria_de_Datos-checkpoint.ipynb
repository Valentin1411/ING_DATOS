 ## Tratando con datos faltantes

Los valores que faltan en un data set, puede darse por varias razones:
extracción de los datos (a la hora de la gestión, recolección de los datos

```python
pd.isnull(datos_csv_url["Club"])
pd.isna(datos_csv_url)
```

```python
#pd.notnull(datos_csv_url["ID"])
pd.notnull(datos_csv_url)
```

```python
#Es importante analizar si hay datos faltantes, por qué puede haber datos faltantes y si los puedo quitar
#datos_csv_url.isna() # verificar en que momento en el campo aparece un booleano true or false
datos_csv_url.isna().sum() #como se puede observar, tenemos datos faltantes.
```

```python
#Podemos remplazarlos por (cero,media/max/mediana por columna, y fila) o simplemente eliminarlas 
#para evitar sesgos. Si son pocos los datos nan o no tienen importancia, se puede eliminarlos sin problemas
#si eliminamos por columnas, pierdo mucha información, pero si eliminamos por filas, pierdo todo, por eso debo agrupar
#qué necesito y ya luego elimino. Si voy a agrupar varias cosas tengo que utilizar doble corchete.
datos_csv_url_clean = datos_csv_url[['ID','Overall','Age','International Reputation','Weak Foot',
                                     'Skill Moves','Name','Club']].dropna(axis = 0, how = 'any')  #drop rows with any missing values in them
print(datos_csv_url_clean.isna().sum())
print("Cantidad de datos iniciales:",len(datos_csv_url))
print("Cantidad de datos resultantes:",len(datos_csv_url_clean))
datos_csv_url_clean
```

## Inspeccionar si hay duplicados

```python
#Primero podemos inspeccionar si hay valores únicos con el siguiente comando:
datos_csv_url_clean.nunique()
#como se puede ver, pese a que hay 5321 filas, solo hay 5182 nombres, eso quiere decir que hay datos repetidos.
```

```python
#Procedemos a inspeccionar primero cuáles podrían ser esos datos repetidos.
#datos_csv_url_clean[datos_csv_url_clean.duplicated(subset = ['Name','Club'],keep=False)].sort_values('Club') #keep=false para que muestre todos los repetidos
#datos_csv_url_clean[datos_csv_url_clean.duplicated(subset = 'Name', keep =False)].sort_values('Name')
#datos_csv_url_clean[datos_csv_url_clean.duplicated(subset = 'Club', keep =False)].sort_values('Club')
datos_csv_url_clean[datos_csv_url_clean.duplicated(subset = ['Name','Club'], keep =False)].sort_values('Name')
```

```python
#si deseo saber cuantas veces se repite un dato:
repeat_names = datos_csv_url_clean.groupby(by = 'Name').size().sort_values(ascending =False)
repeat_names
```

```python
#Pero para saber cuantos nombres y clubes se repiten:
print(len(datos_csv_url_clean))
datos_csv_url_clean.groupby(by = ['Name','Club']).size().sort_values(ascending =False)

#print(len(datos_csv_url_clean))
```

```python
#Si queremos eliminar los duplicados, aplicamos la función drop_duplicates:

datos_csv_url_clean.drop_duplicates(subset=['Name', 'Club'], keep='first', inplace=True)
print(len(datos_csv_url_clean))
datos_csv_url_clean
```

```python
#Verificamos que ya no estén repetidos:
datos_csv_url_clean[datos_csv_url_clean['Name'] == 'Fernandinho'].sort_values('Club')
```

```python
#Verificamos que ya no estén repetidos:
datos_csv_url_clean[datos_csv_url_clean['Name'] == 'A. Miranchuk'].sort_values('Club')
```

```python
#Es posible que para entrenar un modelo necesitemos transformar datos categóricos, como por ejemplo el club.
#Primero contamos cuantas categorías podríamos tener, en este caso son muchas, pero en otros casos puede ser manejable
datos_csv_url_clean['Club'].value_counts()
```

## Transformando datos

```python
#Primero debemos cambiar el tipo de dato a un tipo categórico para que la función logre entender la transformación
datos_csv_url_clean['Club'] = datos_csv_url_clean['Club'].astype('category')
datos_csv_url_clean.dtypes
```

```python
#Lo siguiente es agregar una nueva columna con los datos categorizados gracias a la función cat.codes de pandas
datos_csv_url_clean['Club_cat'] = datos_csv_url_clean['Club'].cat.codes
datos_csv_url_clean.tail() # head muestra el encabezado y tail la parte final. 
#La columna Club_cat ya puede ser la variable que le vamos a entregar al modelo.
```

```python
#Comprobamos que se haya hecho la cotegorización:
datos_csv_url_clean[datos_csv_url_clean['Club_cat'] == 1]
#datos_csv_url_clean['Club_cat'] == 1
```

## Trabajando con datos sin procesar y obteniendo nuevos resultados

```python
datos_csv_url_clean['Test'] = datos_csv_url_clean['Overall']*2
datos_csv_url_clean.head()
#Existen muchas otras operaciones que iremos viendo más adelante
```

```python
#Si queremos agregar una lógica de cuales jugadores son buenos y cuales son malos
datos_csv_url_clean['test2'] = datos_csv_url_clean['Overall'] > 80
#datos_csv_url_clean.tail()
datos_csv_url_clean.head()
```

```python
#Si queremos hacer algo más elaborado, primero creamos una función que cataloga si es bueno o malo
def clasificacion(x):
    if x['Overall'] > 80:        
        return('Bueno')
    else:
        return('Regular')
    
datos_csv_url_clean['clase_jugador'] = datos_csv_url_clean.apply(lambda x: clasificacion(x), axis=1)
datos_csv_url_clean.head()
```

## Escribiendo datos procesados

```python
#Para escribir datos o guardar nuestro dataframe transformado basta con usar: 
# df.to_csv(r'Path where you want to store the exported CSV file\File Name.csv')
datos_csv_url_clean.to_csv('Datos_procesados_soccer.csv', encoding='utf-8-sig')
```
